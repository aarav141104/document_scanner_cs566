```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os

# --- Paths ---
input_path = "data/input1.jpg"
output_path_edges = "results/sample_edges.jpeg"

# --- Step 1: Read the image ---
img = cv2.imread(input_path)

# Convert from BGR (OpenCV default) to RGB (for matplotlib)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# --- Step 2: Convert to grayscale ---
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# --- Step 3: Apply Canny Edge Detection ---
edges = cv2.Canny(gray, 50, 150)

# --- Step 4: Save the edge-detected image ---
cv2.imwrite(output_path_edges, edges)

# --- Step 5: Display both side by side ---
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detected Image")
plt.axis("off")

plt.show()
```

```python
# --- Step 6: Preprocessing for contour detection ---
# Denoise edges slightly
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Equalize histogram to boost contrast
eq = cv2.equalizeHist(blurred)

# Recompute edges after preprocessing
edges_clean = cv2.Canny(eq, 50, 150)
```

```python
# --- Step 7: Find contours ---
contours, _ = cv2.findContours(edges_clean, cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)

# Sort contours by area (largest first)
contours = sorted(contours, key=cv2.contourArea, reverse=True)
doc_contour = None

# Loop over contours to find a 4-point polygon
for cnt in contours:
    peri = cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, 0.02 * peri, True)

    if len(approx) == 4:     # found the document
        doc_contour = approx
        break

if doc_contour is None:
    print(" Document contour not found!")
else:
    print(" Document contour detected!")
```

```python
# --- Step 8: Order the four points (TL, TR, BR, BL) ---
def order_points(pts):
    pts = pts.reshape(4, 2)
    ordered = np.zeros((4, 2), dtype="float32")

    s = pts.sum(axis=1)
    diff = np.diff(pts, axis=1)

    ordered[0] = pts[np.argmin(s)]      # top-left
    ordered[2] = pts[np.argmax(s)]      # bottom-right
    ordered[1] = pts[np.argmin(diff)]   # top-right
    ordered[3] = pts[np.argmax(diff)]   # bottom-left

    return ordered

# Apply ordering
pts = order_points(doc_contour)
```

```python
# --- Step 9: Compute width and height of the new image ---
(tl, tr, br, bl) = pts

# compute width
widthA = np.linalg.norm(br - bl)
widthB = np.linalg.norm(tr - tl)
maxWidth = int(max(widthA, widthB))

# compute height
heightA = np.linalg.norm(tr - br)
heightB = np.linalg.norm(tl - bl)
maxHeight = int(max(heightA, heightB))

# Destination points for the warp
dst = np.array([
    [0, 0],
    [maxWidth - 1, 0],
    [maxWidth - 1, maxHeight - 1],
    [0, maxHeight - 1]
], dtype="float32")

# Compute homography
M = cv2.getPerspectiveTransform(pts, dst)

# Warp the original image
warped = cv2.warpPerspective(img, M, (maxWidth, maxHeight))
```

```python
# --- Step 10: Display everything ---
plt.figure(figsize=(15,6))

plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(edges_clean, cmap='gray')
plt.title("Processed Edges")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(cv2.cvtColor(warped, cv2.COLOR_BGR2RGB))
plt.title("Warped / Scanned Document")
plt.axis("off")

plt.show()

# Save final result
cv2.imwrite("results/scanned_output.jpg", warped)
print(" Saved scanned_output.jpg!")
```
